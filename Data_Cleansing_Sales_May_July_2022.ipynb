### May

```python
df_may.head()
```

```python
df_may.isna().sum()
```

```python
df_may['Delivery Date'].unique()
```

```python
# Membuat salinan data untuk mencegah perubahan pada DataFrame asli
df_copy = df_may.copy()

# Mengganti nilai '-' dengan NaN pada kolom 'Delivery Date'
df_copy['Delivery Date'].replace('-', pd.NaT, inplace=True)

# Mengubah kolom 'Delivery Date' menjadi tipe data datetime
df_copy['Delivery Date'] = pd.to_datetime(df_copy['Delivery Date'])

# Membuat histogram plot untuk kolom 'Delivery Date'
plt.figure(figsize=(10, 6))
plt.hist(df_copy['Delivery Date'].dropna(), bins=30, color='skyblue') # drop NaN value
plt.title('Histogram of Delivery Dates')
plt.xlabel('Delivery Dates')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Mengisi missing value pada kolom Delivery Date dengan nilai rata-rata
df_may['Delivery Date'].replace('-', pd.NaT, inplace=True)
mean_may = df_may['Delivery Date'].mean()
df_may['Delivery Date'].fillna(mean_may, inplace=True)
```

```python
mean_may
```

```python
# Membuat histogram plot untuk kolom 'Delivery Date'
plt.figure(figsize=(10, 6))
plt.hist(df_may['Delivery Date'], bins=30, color='skyblue')
plt.title('Histogram of Delivery Dates')
plt.xlabel('Delivery Dates')
plt.ylabel('Frequency')
plt.xticks(rotation=45)  # Mengatur label tanggal agar dapat terbaca dengan baik
plt.tight_layout()
plt.show()
```
